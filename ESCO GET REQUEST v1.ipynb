### ESCO "Top Level Occupations" API output to JSON format

```python
import json
import requests

URL="https://ec.europa.eu/esco/api/resource/related?"

PARAMS = {"language":"en",\
          "uri":"http://data.europa.eu/esco/concept-scheme/occupations",\
          "relation":"hasTopConcept",}
          
result = requests.get(url = URL, params = PARAMS)

data = result.json()
#data
   
# Indent keyword while dumping the
# data decides to what level 
# spaces the user wants.

print(json.dumps(data, indent = 1))
   
# Difference in the spaces 
# near the brackets can be seen

#print(json.dumps(json_object, indent = 3))
```

### ESCO get-request encoding checker
